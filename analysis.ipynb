```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(np.fromiter((int(x) for x in open("data/python-cycle-DEFG-0.dat")), np.int32, count=9000), label="cyclic references (lists containing self)")
ax.plot(np.fromiter((int(x) for x in open("data/python-mut-DEFG-0.dat")), np.int32, count=9000), label="mutable objects (empty lists)")
ax.plot(np.fromiter((int(x) for x in open("data/python-imm100None-DEFG-0.dat")), np.int32, count=9000), label="immutable objects (tuples)")
ax.plot(np.fromiter((int(x) for x in open("data/python-num-DEFG-0.dat")), np.int32, count=9000), c="gray", label="numerical array (no objects)")
ax.set_yscale("log")
ax.set_ylim(100, 2e8)
ax.set_xlim(0, 9000)
ax.set_ylabel("s between 2048 steps")
ax.set_xlabel("steps / 2048")
ax.set_title("Python 3.12.1 garbage collector")
ax.legend(title="object lifespan: 65536  0", loc="upper right")

fig.savefig("gcmeasure-python.pdf")
```
